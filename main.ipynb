```python
import pandas as pd

df_zh_no_header = pd.read_csv('car_UTF8_zh_with_header.csv')
```

```python
# 首先读取并查看所有唯一值
unique_values = {}

# 遍历每一列，收集所有唯一值
for column in df_zh_no_header.columns:
    unique_values[column] = df_zh_no_header[column].dropna().unique().tolist()

# 打印所有需要翻译的值，以便创建字典:
print("需要翻译的内容：")
for column, values in unique_values.items():
    print(f"\n{column}列的唯一值：")
    print(values)
```

中文表格：插入一行表头

```python
# new one
df_zh_with_header = df_zh_no_header
# 添加新的表头
df_zh_with_header.columns = ['省份', '品牌', '日期', '车型', '城市', '上牌数量', '车型级别', '价格区间', '燃料类型']
# print(df_zh_with_header.head())
# 将修改后的DataFrame保存回CSV文件，覆盖原始文件
df_zh_with_header.to_csv('car_UTF8_zh_with_header.csv', index=False)
```

```python
translation_dict = {
    # 省份翻译
    '北京': 'beijing', '重庆': 'chongqing', '上海': 'shanghai', '天津': 'tianjin',
    '安徽': 'anhui', '福建': 'fujian', '甘肃': 'gansu', '贵州': 'guizhou',
    '广西': 'guangxi', '广东': 'guangdong', '海南': 'hainan', '河北': 'hebei',
    '黑龙江': 'heilongjiang', '湖北': 'hubei', '湖南': 'hunan', '河南': 'henan',
    '江西': 'jiangxi', '江苏': 'jiangsu', '吉林': 'jilin', '辽宁': 'liaoning',
    '内蒙古': 'inner mongolia', '宁夏': 'ningxia', '青海': 'qinghai', '四川': 'sichuan',
    '陕西': 'shaanxi', '山西': 'shanxi', '山东': 'shandong', '西藏': 'tibet',
    '新疆': 'xinjiang', '云南': 'yunnan', '浙江': 'zhejiang',

    # 车型翻译
    'SUV': 'suv',
    '中型车': 'mid-size car',
    '紧凑型车': 'compact car',
    'MPV': 'mpv',
    '中大型车': 'mid-large car',
    '小型车': 'small car',
    '大型车': 'large car',
    '跑车': 'sports car',
    '微型车': 'mini car',

    # 燃油类型翻译
    '燃油': 'fuel',
    '混动': 'hybrid',
    '纯电': 'electric',

}
# 定义翻译函数
def translate_using_dict(text):
    if pd.isna(text):  # 检查是否为空值
        return text
    text = str(text)
    return translation_dict.get(text, text)  # 如果找不到翻译，返回原文

# 对DataFrame中的指定列应用翻译
df_en_with_header = df_zh_with_header.copy()

columns_to_translate = ['省份', '品牌', '车型', '城市', '车型级别', '燃料类型']  # 添加需要翻译的列名
for column in columns_to_translate:
    # 先获取新的列名（如果有翻译的话）
    new_column = translation_dict.get(column, column)
    # 翻译列内容
    df_en_with_header[new_column] = df_en_with_header[column].apply(translate_using_dict)
    # 如果列名已经改变，删除原来的列
    if new_column != column:
        df_en_with_header.drop(column, axis=1, inplace=True)
print(df_en_with_header.head(10))
# df_en.to_csv('car_en.csv', index=False)
```
