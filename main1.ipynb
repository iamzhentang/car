# 价格区间划分

```python
import numpy as np

# 假设我们使用四分位数来粗略估计密集区间
def define_price_ranges(data, label):
    q1 = np.percentile(data, 25)
    median = np.median(data)
    q3 = np.percentile(data, 75)
    print(f"{label} - 密集区间: {q1} 到 {q3}")
    print(f"{label} - 稀疏区间: 小于 {q1} 或大于 {q3}")

# 根据上面的代码，提取三个类别价格区间数据
fuel_prices = df_fuel['价格区间'].values
ev_prices = df_ev['价格区间'].values
phev_prices = df_phev['价格区间'].values
erev_prices = df_erev['价格区间'].values

# 计算价格区间
define_price_ranges(fuel_prices, "燃油车型")
define_price_ranges(ev_prices, "纯电车型")
define_price_ranges(phev_prices, "混动车型")
define_price_ranges(erev_prices, "增程车型")
```

## 燃油类型

```python
# 筛选燃油类型且为合资厂商的数据,按品牌分组统计上牌量并排序
df_matched = cudf.from_pandas(df_matched)
fuel_joint_registrations = df_matched[
    (df_matched['燃料类型'] == '燃油') & 
    (df_matched['厂商类型'] == '合资')
].groupby('品牌')['上牌数量'].sum().sort_values(ascending=False)
print("燃油车型合资厂商上牌总量排名:")
print(fuel_joint_registrations)
# 自主厂商
fuel_private_registrations = df_matched[
    (df_matched['燃料类型'] == '燃油') & 
    (df_matched['厂商类型'] == '自主')
].groupby('品牌')['上牌数量'].sum().sort_values(ascending=False)
print("燃油车型自主品牌上牌总量排名:")
print(fuel_private_registrations)
# 进口厂商
fuel_international_registrations = df_matched[
    (df_matched['燃料类型'] == '燃油') & 
    (df_matched['厂商类型'] == '进口')
].groupby('品牌')['上牌数量'].sum().sort_values(ascending=False)
print("燃油车型进口厂商上牌总量排名:")
print(fuel_international_registrations)
# 造车新势力
fuel_new_registrations = df_matched[
    (df_matched['燃料类型'] == '燃油') & 
    (df_matched['品牌分类'] == '造车新势力')
].groupby('品牌')['上牌数量'].sum().sort_values(ascending=False)
print("燃油车型造车新势力上牌总量排名:")
print(fuel_new_registrations)

# 转换 Series 为 DataFrame
fuel_joint_registrations_df = fuel_joint_registrations.reset_index()
fuel_private_registrations_df = fuel_private_registrations.reset_index()
fuel_international_registrations_df = fuel_international_registrations.reset_index()
fuel_new_registrations_df = fuel_new_registrations.reset_index()

# 保存为 CSV 文件
# fuel_joint_registrations_df.to_csv('rank/fuel_joint_registrations.csv', encoding='utf-8', index=False)
# fuel_private_registrations_df.to_csv('rank/fuel_private_registrations.csv', encoding='utf-8', index=False)
# fuel_international_registrations_df.to_csv('rank/fuel_international_registrations.csv', encoding='utf-8', index=False)
# fuel_new_registrations_df.to_csv('rank/fuel_new_registrations.csv', encoding='utf-8', index=False)
```

# 所有省份

```python
import pandas as pd

# 假设 df_zh_cleaned_zero 是已经清理过的数据
provice_list = {
        '北京': 'beijing', '重庆': 'chongqing', '上海': 'shanghai', '天津': 'tianjin',
    '安徽': 'anhui', '福建': 'fujian', '甘肃': 'gansu', '贵州': 'guizhou',
    '广西': 'guangxi', '广东': 'guangdong', '海南': 'hainan', '河北': 'hebei',
    '黑龙江': 'heilongjiang', '湖北': 'hubei', '湖南': 'hunan', '河南': 'henan',
    '江西': 'jiangxi', '江苏': 'jiangsu', '吉林': 'jilin', '辽宁': 'liaoning',
    '内蒙古': 'inner mongolia', '宁夏': 'ningxia', '青海': 'qinghai', '四川': 'sichuan',
    '陕西': 'shaanxi', '山西': 'shanxi', '山东': 'shandong', '西藏': 'tibet',
    '新疆': 'xinjiang', '云南': 'yunnan', '浙江': 'zhejiang'
}
# 去掉2024-10月份的行
df_zh_cleaned_zero = df_zh_cleaned_zero[df_zh_cleaned_zero['日期'] != '2024-10-01']

for province_zh, province_en in provice_list.items():
    df_province = df_zh_cleaned_zero[df_zh_cleaned_zero['省份'] == province_zh]
    # print(province_en)

    '''按城市、日期和车型级别分组，并计算每组的上牌数量总和'''
    grouped_df0 = df_province.groupby(['城市', '日期', '车型级别']).agg({'上牌数量': 'sum'}).reset_index()

    # 使用 pivot_table 将数据转换为透视表，每列表示一个车型级别
    pivot_table0 = grouped_df0.pivot_table(index=['城市', '日期'], columns=['车型级别'], values='上牌数量', fill_value=0)

    # 查看结果
    # print(pivot_table0)

    '''按城市、日期和燃油类型分组，并计算每组的上牌数量总和'''
    grouped_df1 = df_province.groupby(['城市', '日期', '燃料类型']).agg({'上牌数量': 'sum'}).reset_index()

    # 使用 pivot_table 将数据转换为透视表，每列表示一个燃油类型
    pivot_table1 = grouped_df1.pivot_table(index=['城市', '日期'], columns=['燃料类型'], values='上牌数量', fill_value=0)
    # 查看结果
    # print(pivot_table1)

    # 定义新的价格区间
    price_bins = [0, 5, 10, 15, 20, 25, 30, 35, 40, 50, 100, 150, 300, 1000]
    price_labels = ['0-5w', '5-10w', '10-15w', '15-20w', '20-25w', '25-30w', 
                    '30-35w', '35-40w', '40-50w', '50-100w', '100-150w', '150-300w', '300-1000w']

    # 使用pd.cut创建一个新的列'价格区间标签', 该列将当前价格分配到新的区间
    df_province['价格区间标签'] = pd.cut(df_province['价格区间'], bins=price_bins, labels=price_labels, right=False)

    '''按城市、日期和新价格区间分组，并计算每组的上牌数量总和'''
    grouped_df2 = df_province.groupby(['城市', '日期', '价格区间标签']).agg({'上牌数量': 'sum'}).reset_index()

    # 使用 pivot_table 将数据转换为透视表，每列表示一个新价格区间
    pivot_table2 = grouped_df2.pivot_table(index=['城市', '日期'], columns='价格区间标签', values='上牌数量', fill_value=0)

    # 查看结果
    # print(pivot_table2)

    '''按城市、日期、厂商类型分组，并计算每组的上牌数量总和'''
    grouped_df3 = df_province.groupby(['城市', '日期', '厂商类型']).agg({'上牌数量': 'sum'}).reset_index()

    # 使用 pivot_table 将数据转换为透视表，每列表示一个厂商类型
    pivot_table3 = grouped_df3.pivot_table(index=['城市', '日期'], columns=['厂商类型'], values='上牌数量', fill_value=0)

    # 查看结果
    # print(pivot_table3)

    '''按城市、日期和品牌分类分组，并计算每组的上牌数量总和'''
    grouped_df4 = df_province.groupby(['城市', '日期', '品牌分类']).agg({'上牌数量': 'sum'}).reset_index()

    # 使用 pivot_table 将数据转换为透视表，每列表示一个品牌分类
    pivot_table4 = grouped_df4.pivot_table(index=['城市', '日期'], columns=['品牌分类'], values='上牌数量', fill_value=0)

    # 查看结果
    # print(pivot_table4)


    # 确保所有的索引顺序一致
    pivot_table0 = pivot_table0.sort_index()
    pivot_table1 = pivot_table1.sort_index()
    pivot_table2 = pivot_table2.sort_index()
    pivot_table3 = pivot_table3.sort_index()
    pivot_table4 = pivot_table4.sort_index()
    # 拼接上牌数量、燃料类型和价格区间的透视表
    result = pd.concat([pivot_table0, pivot_table1, pivot_table2, pivot_table3, pivot_table4], axis=1)

    # 重置索引，使得'城市'和'日期'成为列
    result.reset_index(inplace=True)

    # 查看结果
    # print(result.head())
    # 生成csv前需要用户确认
    # result.to_csv(f'result/result_{province_en}.csv', encoding='gbk', index=False)
```
